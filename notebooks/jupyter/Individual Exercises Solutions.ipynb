## Transformations and Actions

### Aggregations

Find the maximum QUANTITY value in the transaction_data dataframe.

```
import pyspark.sql.functions as f
transactions.select(f.max('QUANTITY')).display()
```

What values are in the MARITAL_STATUS_CODE column in the hh_demographics dataframe?

```
hh_demographic.select('MARITAL_STATUS_CODE').distinct().display()
```

### Filtering

Filter the product dataframe so that it only has department values of 'PRODUCE'.

```
product.filter(f.col('DEPARTMENT') == 'PRODUCE').display()
```

### Joining

Join the transaction_data and hh_demographic dataframes.

```
txn_hh = transactions.join(hh_demographic, on='household_key')
txn_hh.display()
```

Join the product dataframe to the above dataframe.

```
txn_hh_prod = txn_hh.join(product, on='PRODUCT_ID')
txn_hh_prod.display()
```

What is the average sales value for households in the 25-34 age range?

```
age_filt = txn_hh_prod.filter(f.col('AGE_DESC') == '25-34')
age_filt.agg(f.avg('SALES_VALUE')).display()
```

What percentage of products bought by households in the 25-34 age range are in the PRODUCE department?

```
age_filt.filter(f.col('DEPARTMENT') == 'PRODUCE').count()/age_filt.count()*100
```
